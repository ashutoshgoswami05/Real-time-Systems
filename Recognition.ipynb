```python
from keras.models import load_model
import cv2
import numpy as np

namedict = {0: "Ashutosh", 1: "Bill Gates",2:"Elon Musk",3:"Jeff Bezoz",4:"Mark ZuckerBerg",5:"Steve Jobs",6:"Nand",7:"Tallat"}
np.set_printoptions(suppress=True)

# here we are loading model we saved in previous step 
model = load_model("keras_Model.h5", compile=False)
class_names = open("labels.txt", "r").readlines()
captureWindow = cv2.VideoCapture(0)

cascadeFace = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

while True:
    ret, image = captureWindow.read()
    displayImage = image.copy()

    # Convert the image to grayscale for face detection
    grayImage = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Perform face detection
    faces = cascadeFace.detectMultiScale(grayImage, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    for (x, y, w, h) in faces:
        cv2.rectangle(displayImage, (x, y), (x + w, y + h), (255, 0, 0), 2)

        capturedFace = image[y:y + h, x:x + w]

         # Resizing the face input to (224, 224) as it was input size to the vgg model
        capturedFace = cv2.resize(capturedFace, (224, 224), interpolation=cv2.INTER_AREA)
        capturedFace = np.asarray(capturedFace, dtype=np.float32).reshape(1, 224, 224, 3)

        # Normalize the capturedFace array
        capturedFace = (capturedFace / 127.5) - 1

        # Prediction using the model
        prediction = model.predict(capturedFace)
        index = np.argmax(prediction)
        class_name = class_names[index]
        confidence_score = prediction[0][index]

        print("Class:", class_name[2:], end="")
        print("Confidence Score:", str(np.round(confidence_score * 100))[:-2], "%")

        if index in namedict:
            if confidence_score>=0.90:    # if confidence score is greater than 90% then only we show the corresponding name of the person
                nametoput = namedict[index]
                cv2.putText(displayImage, nametoput, (50, 50), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 255, 0), 2)
            else:
                 cv2.putText(displayImage, "Unknown", (50, 50), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 255, 0), 2)
        else:
            cv2.putText(displayImage, "No face found", (50, 50), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 255, 0), 2)

    cv2.imshow("Webcam Image", displayImage)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

captureWindow.release()
cv2.destroyAllWindows()
```
